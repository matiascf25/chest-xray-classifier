```python
# Importación de librerías a utilizar
import os
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm

# Para Deep Learning
import torch
from torchvision import transforms, datasets
```

```python
# Para visualizar cantidad de imagenes en cada clase y subset
data_dir = 'data/chest_xray'
for split in ['train', 'val', 'test']:
    split_path = os.path.join(data_dir, split)
    if not os.path.exists(split_path):
        print(f"{split} no existe.")
        continue
    print(f"{split} folders:")
    for cls in os.listdir(split_path):
        n_imgs = len(os.listdir(os.path.join(split_path, cls)))
        print(f"  {cls}: {n_imgs} imágenes")
```

```python
# Visualizar algunas de las imagenes del DataSet
for split in ["train", "test"]:
    for cls in os.listdir(os.path.join(data_dir, split)):
        folder = os.path.join (data_dir, split, cls)
        file = os.listdir(folder)[0]
        img= Image.open(os.path.join(folder,file))
        plt.imshow(img, cmap="gray")
        plt.title(f"{split} - {cls}")
        plt.axis ("off")
        plt.show()
```

```python
# Definir transformaciones para las imágenes
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.Grayscale(num_output_channels=3),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
# Crear DataSets
train_dataset = datasets.ImageFolder(root="data/chest_xray/train", transform=transform)
test_dataset = datasets.ImageFolder(root="data/chest_xray/test", transform=transform)
```

```python
# Crear DataLoaders
from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
# Comprobar los DataLoaders
batch = next(iter(train_loader))
images, labels = batch
print("Shape imágenes:", images.shape)
print("Labels:", labels)
```

```python
from torch.utils.data import random_split

# 80% train, 20% val
train_size =int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size

train_ds, val_ds = random_split(train_dataset, [train_size, val_size])

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
val_loader =DataLoader(val_ds, batch_size=32, shuffle=False)
```

```python
# Evaluación General del Dataset
# Construimos un DataFrame con info relevante
data_dir = "data/chest_xray/train"
# Lista para almacenar la info
rows = []

for cls in os.listdir(data_dir):
    class_dir = os.path.join(data_dir, cls)
    if not os.path.isdir(class_dir):
        continue
    for fname in tqdm(os.listdir(class_dir), desc=cls):
        fpath = os.path.join(class_dir, fname)
        try:
            img = Image.open(fpath)
            arr = np.array(img)
            rows.append({
                "filename": fname,
                "class": cls,
                "width": img.width,
                "height": img.height,
                "mean_pixel": arr.mean(),
                "std_pixel": arr.std()
            })
        except Exception as e:
            print(f"Error leyendo {fpath}: {e}")

# Crear DataFrame
df = pd.DataFrame(rows)

# Vista previa
print(df.head())
print(df.info())
```

Columnas: filename, class, width, height, mean_pixel, std_pixel

Valores faltantes: es raro en metadatos de imágenes, pero puedes revisar si falta alguna estadística.

Justificación: Generar este DataFrame es esencial para análisis exploratorio en datasets de imágenes, pues te permite descubrir errores, imágenes corruptas o anomalías de tamaño.

```python
# Eliminar filas con valores nulos
df = df.dropna()
```

```python
# Normalización
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[["mean_pixel_norm", "std_pixel_norm"]] = scaler.fit_transform(df[["mean_pixel", "std_pixel"]])
```

```python
# Análisis descriptivo
print(df.describe())
```

```python
# Distribución de clases
print(df["class"].value_counts())
```

```python
# Comparativa por clases
print(df.groupby("class")[["mean_pixel", "std_pixel"]].describe())
```

```python
# Hisograma de media de intensidad por clase
for c in df["class"].unique():
    plt.hist(df[df["class"]==c]["mean_pixel"], bins=30, alpha=0.6, label=c)
plt.xlabel("Intensidad media (0-255)")
plt.ylabel("Número de imágenes")
plt.title("Distribución de intensidad media por clase")
plt.legend()
plt.show()
```

```python
# Boxplot de tamaño de imagen por clases
import seaborn as sns

sns.boxplot(x="class", y="width", data=df)
plt.title("Distribución del ancho de imagen por clase")
plt.show()
```

```python
# Gráfico de dispersión entre media y std
sns.scatterplot(x="mean_pixel", y="std_pixel", hue="class", data=df, alpha=0.5)
plt.title("Relación entre brillo medio y desviación estándar por clase")
plt.xlabel("Brillo medio")
plt.ylabel("Desviación estándar")
plt.show()
```
